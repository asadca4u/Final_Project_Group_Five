```python
#Build Summary Tables
# Accuracy
data_subset = ["All Data", "Male Data", "Female Data", "Under 40", "40-60", "Over 60"]
Acc_table = pd.DataFrame(index=data_subset, data=([[0.915, 0.954, 0.985, 0.985], 
                                                   [0.939, 0.939, 0.951, 1.0],
                                                   [0.938, 1.0, 0.979, 0.979],
                                                   [0.861, 0.944, 0.972, 0.972],
                                                   [0.972, 0.930, 0.986, 1.0], 
                                                   [1.0, 1.0, 0.917, 0.958]]),
    columns=(["LR", "SVM", "DT", "RF"]))
Acc_table.loc['Mean'] = Acc_table.mean()
Acc_table = Acc_table.round(decimals=3).astype(object)
Acc_table
```

```python
# Precision
data_subset = ["All Data", "Male Data", "Female Data", "Under 40", "40-60", "Over 60"]
Pre_table = pd.DataFrame(index=data_subset, data=([[0.92, 0.95, 0.98, 0.98], 
                                                   [0.94, 0.94, 0.95, 1.0],
                                                   [0.94, 1.0, 0.98, 0.98],
                                                   [0.88, 0.95, 0.97, 0.97],
                                                   [0.97, 0.93, 0.99, 1.0], 
                                                   [1.0, 1.0, 0.93, 0.96]]),
    columns=(["LR", "SVM", "DT", "RF"]))
Pre_table.loc['Mean'] = Pre_table.mean()
Pre_table = Pre_table.round(decimals=2).astype(object)
Pre_table
```

```python
# Sensitivity
data_subset = ["All Data", "Male Data", "Female Data", "Under 40", "40-60", "Over 60"]
Sen_table = pd.DataFrame(index=data_subset, data=([[0.92, 0.95, 0.98, 0.98], 
                                                   [0.94, 0.94, 0.95, 1.0],
                                                   [0.94, 1.0, 0.98, 0.98],
                                                   [0.86, 0.94, 0.97, 0.97],
                                                   [0.97, 0.93, 0.99, 1.0], 
                                                   [1.0, 1.0, 0.92, 0.96]]),
    columns=(["LR", "SVM", "DT", "RF"]))
Sen_table.loc['Mean'] = Sen_table.mean()
Sen_table = Sen_table.round(decimals=2).astype(object)
Sen_table
```

```python
# Plot Accuracy
import plotly.graph_objects as go
data_subset = ["All Data", "Male Data", "Female Data", "Under 40", "40-60", "Over 60", "Mean"]
fig = go.Figure(data=[
    go.Bar(name='Linear Regression', x=data_subset, y=Acc_table["Linear Regression"]),
    go.Bar(name='SVM', x=data_subset, y=Acc_table["SVM"]),
    go.Bar(name='Decision Tree', x=data_subset, y=Acc_table["Decision Tree"]),
    go.Bar(name='Random Forest', x=data_subset, y=Acc_table["Random Forest"])

])
# Change the bar mode
fig.update_layout(
    barmode='group',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Accuracy',
        titlefont_size=16,
        tickfont_size=14))

fig.show()
```
